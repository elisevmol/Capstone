**Finding Number of Topics**

```python
complete_vocab_set = set(complete_vocab)

stripped_abstracts = []
for abstract in complete_abstracts:
    stripped_abstracts.append(strip_abstract(abstract, complete_vocab_set))
    
dictionary = corpora.Dictionary(stripped_abstracts)

corpus = []
for text in stripped_abstracts:
    corpus.append(dictionary.doc2bow(text))
```

```python
all_topic_coherence = []
for i in tqdm_notebook(range(100)[1:]):
    all_topic_coherence.append(find_nt(i, corpus, dictionary))
```

```python
def convert_to_mean_and_std(one_topic_coherence):
    the_list = [float(i.split(']')[0][1:]) for i in one_topic_coherence[0][:-3]]
    mean = np.mean(the_list)
    std_dev = np.std(the_list)
    return [mean, std_dev]

mean_and_std_for_all = [convert_to_mean_and_std(x) for x in all_topic_coherence]

y = [x[0] for x in mean_and_std_for_all]
std = [x[1] for x in mean_and_std_for_all]
x = range(99)

plt.errorbar(x, y, std, color='black', linestyle='-', marker='o', markersize=4, 
             markerfacecolor='black', elinewidth=0.5, markeredgecolor='black')

plt.title('Physics - Number of Topics', fontsize=12)
plt.ylabel('Average Topic Coherence', fontsize=10)
plt.xlabel('Number of Topics', fontsize=10)
plt.ylim([0, 0.06])
plt.xlim([20, 100])
plt.grid(True)
plt.yticks(fontsize = 7, weight='light')
plt.xticks(fontsize = 7, weight='light')

plt.savefig('/Users/elise/Documents/Capstone/Graphs/TC-physics.png')
plt.show()

# small range
y = [x[0] for x in mean_and_std_for_all]
std = [x[1] for x in mean_and_std_for_all]
x = range(99)

plt.errorbar(x, y, std, color='black', linestyle='-', marker='o', markersize=4, 
             markerfacecolor='black', elinewidth=0.5, markeredgecolor='black')

plt.title('Physics - Number of Topics', fontsize=12)
plt.ylabel('Average Topic Coherence', fontsize=10)
plt.xlabel('Number of Topics', fontsize=10)
plt.ylim([0, 0.04])
plt.xlim([40, 60])
plt.grid(True)
plt.yticks(fontsize = 7, weight='light')
plt.xticks(fontsize = 7, weight='light')

plt.savefig('/Users/elise/Documents/Capstone/Graphs/TC-physics-40_50.png')
plt.show()
```

```python
cluster1 = [32, 14, 25, 28, 0, 20, 13, 36, 30, 31, 3, 38, 7, 37, 6, 34]
cluster3 = [11, 8, 33, 16, 12, 24, 5, 23]
cluster4 = [18, 29] 
cluster5 = [22, 1, 9, 19, 26, 27, 4, 15]
cluster6 = [10, 17, 2, 35, 39]
```

```python
def plot_average(the_data, numbers):
    
    time_series = zip(*[the_data[n] for n in numbers])

    mean_std = []
    for years in time_series:
        y = [year[1] for year in years]
        mean_std.append([years[0][0], np.mean(y), np.std(y)])
     
    fig = plt.figure(1)
    ax = fig.add_subplot(111, axisbg='black')
    
    std = [x[2] for x in mean_std]
    y = [x[1] for x in mean_std]
    x = [int(x[0]) for x in mean_std]
    
    plt.errorbar(x, y, std, color='azure', linestyle='-', marker='o', markersize=4, 
                 markerfacecolor='azure', elinewidth=0.5, ecolor='azure')
    
    plt.gca().set_ylim(bottom=0)
    plt.title('Average for Cluster')
    plt.show()
```

```python
plot_average(the_data, cluster1)
plot_average(the_data, cluster3)
plot_average(the_data, cluster4)
plot_average(the_data, cluster5)
plot_average(the_data, cluster6)
```
